```python
import requests
import math
```

```python
auth=('admin','supersecret123')
limit=100

r = requests.get("http://localhost:5000/api/news?linked=true", auth=auth)
result = r.json()

linked=[]
for i in range(math.ceil(result['total_count']/limit)):
    print("Getting page {}".format(i+1))
    offset = i*limit
    
    r = requests.get("http://localhost:5000/api/news",
                     params={"linked":"true", "limit":limit,
                            "offset":offset},
                     auth=auth)
    linked.extend(r.json()['articles'])
```

```python
def get_candidates(article):
    r = requests.get("http://localhost:5000/api/news/{}/candidates".format(article['id']), auth=auth)
    result = r.json()
                                                                                          
    return result['candidate']['doi_paper']

get_candidates({"id":40})

```

```python
def get_links(article):
    r = requests.get("http://localhost:5000/api/news/{}/links".format(article['id']), auth=auth)
    result = r.json()
    
    return result
    
get_links({"id":40})
```

```python
total = 0
top_1 = 0
top_3 = 0
top_5 = 0

cand_count = []

for article in linked:
    print(article['id'])
    links = get_links(article)
    
    candidates = sorted(get_candidates(article).items(), key=lambda x: x[1], reverse=True)

    cand_count.append(len(candidates))

    sorted_dois = [x[0] for x in candidates]

    for link in links:

        if link['doi'] in sorted_dois[0:1]:
            top_1 += 1

        if link['doi'] in sorted_dois[0:3]:
            top_3 += 1

        if link['doi'] in sorted_dois[0:5]:
            top_5 += 1

        total += 1
```

```python
print("Total:", total)
print("R@1:", top_1)
print("R@3:", top_3)
print("R@5:", top_5)
```
